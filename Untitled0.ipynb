```python
# PASO 1: Carga del conjunto de datos

import os

# 1) Ruta correcta a la carpeta con las imágenes
ruta_train = "/content/data/dogs-vs-cats/train"

# 2) Comprobamos que la carpeta existe
if os.path.exists(ruta_train):
    print("✅ La carpeta 'train' se encontró correctamente.")
else:
    print("❌ No se encontró la carpeta. Revisa la ruta.")

# 3) Listamos todos los archivos que hay
archivos = os.listdir(ruta_train)
print("Cantidad total de archivos encontrados:", len(archivos))

# 4) Vemos los primeros 5 nombres
print("\nEjemplos de nombres de archivo:")
for nombre in archivos[:5]:
    print(" -", nombre)

# 5) Contamos cuántos son perros y cuántos son gatos

perros = 0
gatos = 0
for nombre in archivos:
    if nombre.startswith("dog"):
        perros += 1
    elif nombre.startswith("cat"):
        gatos += 1

print("\n🐶 Perros:", perros)
print("🐱 Gatos:", gatos)
```

```python
# PASO 2 (parte 1): Visualizar las imágenes

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

ruta_train = "/content/data/dogs-vs-cats/train"

# Separamos los nombres de archivo de perros y de gatos
nombres_perros = [n for n in os.listdir(ruta_train) if n.startswith("dog")]
nombres_gatos  = [n for n in os.listdir(ruta_train) if n.startswith("cat")]

# ---------- 9 PERROS ----------
plt.figure(figsize=(8, 8))
for i in range(9):
    imagen = mpimg.imread(os.path.join(ruta_train, nombres_perros[i]))
    plt.subplot(3, 3, i + 1)
    plt.imshow(imagen)
    plt.axis("off")
plt.suptitle("Primeras 9 fotos de PERROS 🐶", fontsize=18, color="#FF6B35", fontweight="bold")
plt.tight_layout()
plt.show()

# ---------- 9 GATOS ----------
plt.figure(figsize=(8, 8))
for i in range(9):
    imagen = mpimg.imread(os.path.join(ruta_train, nombres_gatos[i]))
    plt.subplot(3, 3, i + 1)
    plt.imshow(imagen)
    plt.axis("off")
plt.suptitle("Primeras 9 fotos de GATOS 🐱", fontsize=18, color="#8E44FF", fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# PASO 2 (parte 2): Organizar imágenes en carpetas

import os
import shutil

ruta_origen = "/content/data/dogs-vs-cats/train"

# 1) Creamos las 4 carpetas
os.makedirs("/content/dataset/train/dogs", exist_ok=True)
os.makedirs("/content/dataset/train/cats", exist_ok=True)
os.makedirs("/content/dataset/test/dogs", exist_ok=True)
os.makedirs("/content/dataset/test/cats", exist_ok=True)

# 2) Listamos todos los archivos de la carpeta original
todos = os.listdir(ruta_origen)

# 3) Separamos perros y gatos
perros = []
gatos = []
for nombre in todos:
    if nombre.startswith("dog"):
        perros.append(nombre)
    else:
        gatos.append(nombre)

# 4) Los ordenamos
perros.sort()
gatos.sort()

# 5) Calculamos cuántos van a train (80%)
corte_perros = int(len(perros) * 0.8)
corte_gatos = int(len(gatos) * 0.8)

# 6) Copiamos los PERROS (los primeros 80% a train, el resto a test)
contador = 0
for nombre in perros:
    if contador < corte_perros:
        shutil.copy(ruta_origen + "/" + nombre, "/content/dataset/train/dogs/" + nombre)
    else:
        shutil.copy(ruta_origen + "/" + nombre, "/content/dataset/test/dogs/" + nombre)
    contador = contador + 1

# 7) Copiamos los GATOS
contador = 0
for nombre in gatos:
    if contador < corte_gatos:
        shutil.copy(ruta_origen + "/" + nombre, "/content/dataset/train/cats/" + nombre)
    else:
        shutil.copy(ruta_origen + "/" + nombre, "/content/dataset/test/cats/" + nombre)
    contador = contador + 1

print("✅ ¡Imágenes organizadas correctamente!")

# 8) Verificamos cuántas quedaron en cada carpeta
print("train/dogs:", len(os.listdir("/content/dataset/train/dogs")))
print("train/cats:", len(os.listdir("/content/dataset/train/cats")))
print("test/dogs:", len(os.listdir("/content/dataset/test/dogs")))
print("test/cats:", len(os.listdir("/content/dataset/test/cats")))
```

```python
# PASO 2 (parte 3): Crear los generadores

from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1) Creamos un generador para train y otro para test.
generador_train = ImageDataGenerator(rescale=1./255)
generador_test = ImageDataGenerator(rescale=1./255)

# 2) Cargamos las imagenes de la carpeta train
trdata = generador_train.flow_from_directory(
    "/content/dataset/train",
    target_size=(200, 200),
    batch_size=32,
    class_mode="categorical"
)

# 3) Cargamos las imagenes de la carpeta test
tsdata = generador_test.flow_from_directory(
    "/content/dataset/test",
    target_size=(200, 200),
    batch_size=32,
    class_mode="categorical"
)

# 4) Vemos que etiqueta le asigno a cada clase
print("Clases:", trdata.class_indices)
```

```python
# PASO 3 (parte 1): Construir la red neuronal (CNN)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense

# Creamos el modelo vacio
model = Sequential()

# --- Bloque 1 ---
model.add(Conv2D(input_shape=(200,200,3), filters=64, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

# --- Bloque 2 ---
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

# --- Bloque 3 ---
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

# --- Bloque 4 ---
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

# --- Bloque 5 ---
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))

# --- Capas densas (la decision final) ---
model.add(Flatten())
model.add(Dense(units=4096, activation="relu"))
model.add(Dense(units=4096, activation="relu"))
model.add(Dense(units=2, activation="softmax"))

# Mostramos un resumen de la red
model.summary()
```

```python
# PASO 3 (parte 2): Compilar y entrenar el modelo

# 1) Compilamos el modelo
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# 2) Entrenamos el modelo
historial = model.fit(
    trdata,
    validation_data=tsdata,
    epochs=10
)
```

```python
# PASO 4 (parte 1): Optimizar el modelo

from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers import Adam

# 1) ModelCheckpoint
checkpoint = ModelCheckpoint(
    "mejor_modelo.keras",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

# 2) EarlyStopping
early = EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    mode="max"
)

# 3) Compilamos OTRA VEZ
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# 4) Entrenamos de nuevo
historial = model.fit(
    trdata,
    validation_data=tsdata,
    epochs=20,
    callbacks=[checkpoint, early]
)
```

```python
# PASO 4 (parte 2): VGG16 ya entrenado (transfer learning)

from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import Adam

# 1) Cargamos el VGG16 YA entrenado con millones de imagenes

base = VGG16(weights="imagenet", include_top=False, input_shape=(200,200,3))

# 2) "Congelamos" el VGG16 para NO reentrenarlo
base.trainable = False

# 3) Construimos nuestro modelo: el VGG16 entrenado + nuestra parte final
model2 = Sequential()
model2.add(base)
model2.add(Flatten())
model2.add(Dense(units=128, activation="relu"))
model2.add(Dense(units=2, activation="softmax"))

# 4) Compilamos
model2.compile(optimizer=Adam(learning_rate=0.0001),
               loss="categorical_crossentropy",
               metrics=["accuracy"])

# 5) Vemos el resumen
model2.summary()
```

```python
# PASO 4 (parte 3): Entrenar el modelo con transfer learning

from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# ModelCheckpoint: guarda el MEJOR modelo automaticamente
checkpoint = ModelCheckpoint(
    "mejor_modelo.keras",
    monitor="val_accuracy",
    save_best_only=True,
    mode="max"
)

# EarlyStopping
early = EarlyStopping(
    monitor="val_accuracy",
    patience=3,
    mode="max"
)

# Entrenamos el modelo nuevo
historial = model2.fit(
    trdata,
    validation_data=tsdata,
    epochs=10,
    callbacks=[checkpoint, early]
)
```

```python
# PASO 4 (parte 4): Cargar el mejor modelo y predecir

from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt

# 1) Cargamos el mejor modelo
mejor_modelo = load_model("mejor_modelo.keras")

# 2) Tomamos un grupo de imagenes del conjunto de test
imagenes, etiquetas_reales = next(tsdata)

# 3) El modelo predice sobre esas imagenes
predicciones = mejor_modelo.predict(imagenes)

# 4) Nombres de las clases (recordamos: cats=0, dogs=1)
nombres = ["gato", "perro"]

# 5) Mostramos 9 imagenes con la prediccion
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(imagenes[i])

    # argmax nos da la posicion del valor mas alto (la clase elegida)
    clase_predicha = np.argmax(predicciones[i])
    clase_real = np.argmax(etiquetas_reales[i])

    # verde si acerto, rojo si fallo
    if clase_predicha == clase_real:
        color = "#00C9A7"
    else:
        color = "#FF3B3B"

    plt.title("Prediccion: " + nombres[clase_predicha], color=color, fontweight="bold")
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
resultado = mejor_modelo.evaluate(tsdata)
print("Precision final en test:", round(resultado[1] * 100, 2), "%")
```
